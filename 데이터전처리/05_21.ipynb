```python
X = [32,64,96,118,126,144,152,158]
Y = [17,24,62,49,52,105,130,125]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
```

```python
A = np.c_[X,np.ones(len(X))]
print(A)
```

```python
B = np.array(Y)
```

```python
B
```

```python
int_A = np.linalg.inv(A.T@A)@A.T
```

```python
X = int_A @ B
print(X)
```

```python
plt.scatter(X,Y)
plt.plot(X, np.multiply(X))
```

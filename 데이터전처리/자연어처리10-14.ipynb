```python
import requests
from bs4 import BeautifulSoup

def get_naver_news(url):

    res = requests.get(url)
    bs = BeautifulSoup(res.content, 'html.parser')
    
    date = bs.select('span.media_end_head_info_datestamp_time')[0].text
    title = bs.select('h2#title_area')[0].text
    media = bs.select('div.media_end_head_top a img')[0]['title']
    content = bs.select('div.newsct_article')[0].get_text().replace('\t', '').replace('\n','')

    return (date, title, media, content)
```

```python
url = 'https://news.naver.com/main/read.nhn?model=LSD&mid=sec&sid1=105&oid=030&aid=0002700985'

get_naver_news(url)
```

```python
import requests
from bs4 import BeautifulSoup

def get_naver_blog(url):
    res = requests.get(url)
    bs = BeautifulSoup(res.content, 'html.parser')

    new_url = 'https://blog.naver.com/' + bs.select('iframe')[0]['src']
    res = requests.get(new_url)
    bs = BeautifulSoup(res.content, 'html.parser')

    date = bs.select('span.se_publishDate')[0].text
    title = bs.select('div.se-title-text p span')[0].text
    content = bs.select('div.se-main-container')[0].get_text().replace('\n','')

    return(date, title, content)
```

```python
url = 'https://blog.naver.com/biomath2k/221755824489'

get_naver_blog(url)
```

```python
#word_tokenize() : 마침표&구두점(. , ? ; ! 등)으로 구분하여 토큰화
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

text = 'Barack Obama likes fried chicken very much'
word_tokens = word_tokenize(text)
print(word_tokens)
```

```python
#wordPunctTokenizer() : 알파벳이 아닌 문자를 구분하여 토큰화
import nltk
from nltk.tokenize import WordPunctTokenizer

text = 'Barack Obama likes fried chicken very much'
wordpuncttoken = WordPunctTokenizer().tokenize(text)
print(wordpuncttoken)
```

```python
#TreebankWordTokenizer() : 정규표현식에 기반한 토큰화
import nltk
from nltk.tokenize import TreebankWordTokenizer

text = 'Barach Obama likes fried chicken very much'
treebankwordtoken = TreebankWordTokenizer().tokenize(text)
print(treebankwordtoken)
```

```python
#코모란 Komoran 토큰화
from konlpy.tag import Komoran
komoran = Komoran()
kor_text = '인간이 컴퓨터와 대화하고 있다는 것을 깨닫지 못하고 인간과 대화를 계속할 수 있다면 컴퓨터는 지능적인 것으로 간주될 수 있습니다.'
komoran_tokens = komoran.morphs(kor_text)
print(komoran_tokens)
```

```python
#한나눔 Hannanum 토큰화
from konlpy.tag import Hannanum
hannanum = Hannanum()
kor_text = '인간이 컴퓨터와 대화하고 있다는 것을 깨닫지 못하고 인간과 대화를 계속할 수 있다면 컴퓨터는 지능적인 것으로 간주될 수 있습니다.'
hannanum_tokens = hannanum.morphs(kor_text)
print(hannanum_tokens)
```

```python
# Okt 토큰화 (소셜 데이터를 토큰화하기 적합함)
from konlpy.tag import Okt
okt = Okt()
kor_text = '인간이 컴퓨터와 대화하고 있다는 것을 깨닫지 못하고 인간과 대화를 계속할 수 있다면 컴퓨터는 지능적인 것으로 간주될 수 있습니다.'
okt_tokens = okt.morphs(kor_text)
print(okt_tokens)
```

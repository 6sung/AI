```python
import numpy as np
x = np.linspace(0,10,100)
y = np.cos(x)
fig, axes = plt.subplots(1,2)
axes[0].scatter(x,y,marker=".")
axes[0].text(2,0,'Example Graph', style='italic')
axes[1].scatter(x,y,marker="*")
axes[1].annotate("Sine", xy=(5,0.5), xytext=(2, 0.75),arrowprops=dict(arrowstyle="->", connectionstyle="angle3"))
plt.show()
```

```python
plt.scatter(x,y, marker=".")
plt.text(3,0.5, r"$\sum_{i=0}^\infty X_i$", fontsize=20)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(8,5))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
axes[0,0].scatter(x,y, marker=".")
axes[0,0].set(title="An Example Axes", ylabel="Y-Axis", xlabel="X-Axis")
axes[0,1].scatter(x,y, marker="^", c="r")
axes[0,1].set_xlim(0,5)
axes[0,1].set_ylim(-2,2)
axes[0,1].set_xlabel("X 0~5")
axes[0,1].set_ylabel("Y -2~2")

axes[1,0].scatter(x,y,marker=".",c="c")
axes[1,0].set(xticks=range(1,8,2), xticklabels=[3,100,-12,"foo"], yticks=[-2,0,1,2], yticklabels=[-200,0,100,"Max"])

axes[1,1].scatter(x,y,marker="v",c="b")
axes[1,1].set_xticks(range(1,8,2))
axes[1,1].set_xticklabels([3,100,-12,"foo"])
axes[1,1].set_yticks([-2,0,1,2])
axes[1,1].set_yticklabels([-200,0,100,"Max"])
axes[1,1].spines['top'].set_visible(False)
axes[1,1].spines['bottom'].set_position(("outward",10))
axes[1,1].grid(True)
plt.show()
```

```python
x = np.arange(0,10)
y1 = 0.5 * x**2
y2 = -1 * y1
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x,y1,'g^:')
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='g')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'bv--')
ax2.set_ylabel('Y2 data', color='b')

ax3 = ax1.twiny()
ax3.plot(-x, y1, 'ro-.')
ax3.set_xlabel('-X data', color="r")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 10, 0.6)
y = np.sin(x)
fig, axes = plt.subplots(2,2,figsize=(8,5))
axes[0,0].plot(y)
axes[0,1].plot(x,y,'ro:')
axes[1,0].plot(x,np.cos(x),'cs-.')
axes[1,1].plot([3,6],[3,5],'b^:')
axes[1,1].plot([4,5], [5,4], 'kx--')
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
plt.scatter(x=iris.petal_length, y=iris.petal_width, s=iris.sepal_length*10, c=iris.sepal_width, alpha=0.5)
plt.show()
```

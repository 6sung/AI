```python
from sklearn.datasets import make_classification
X, y = make_classification(n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=2, random_state=9)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], marker='o', c=y, s=100, edgecolor='k', linewidth=2)
plt.show()
```

```python
X[y==1, :] = -(X[y==0, :])
```

```python
plt.scatter(X[:,0], X[:,1], marker='o', s=100, edgecolor='k', linewidth=2)
plt.show()
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
model = QuadraticDiscriminantAnalysis()
model.fit(X,y)
```

```python
x = [[0,0]]
p = model.predict_proba(x)[0]
plt.subplot(211)
plt.scatter(X.T[0], X.T[1], s=10)
plt.scatter(x[0][0], x[0][1], c='k', s=100, marker='x')
plt.title("Test data")
plt.subplot(212)
plt.bar(model.classes_, p)
plt.title("Conditional probability distribution")
plt.xticks(model.classes_, ["$P(y=0|x_{text})$", "$P(y=1|x_{test})$"])
plt.tight_layout()
plt.show()
```

```python
h = model.predict(X)
plt.scatter(X[h==0, 0], X[h==0,1], marker='^', c="r", s=100, edgecolor="k", linewidth=2)
plt.scatter(X[h==1, 0], X[h==1,1], marker='o', c="b", s=100, edgecolor="k", linewidth=2)
plt.show()
```

```python
from sklearn.datasets import make_classification
X, y = make_classification(n_features=1, n_redundant=0, n_informative=1, n_clusters_per_class=1, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X,y)
```

```python
import numpy as np
xx = np.linspace(-3, 3, 100)
XX = xx[:, np.newaxis]
prob = model.predict_proba(XX)[:,1]
```

```python
import matplotlib.pyplot as plt
```

```python
test_X = [[-0.2]]
plt.subplot(211)
plt.plot(xx, prob)
plt.scatter(X,y, marker='o', c=y, s=100, edgecolor='k', linewidth=2)
plt.scatter(test_X[0], model.predict_proba(test_X)[0][1:], marker='x',s=200,c='r',lw=5)
plt.xlim(-3,3)
plt.ylim(-.2,1.2)
plt.legend(["$P(y=1|x_{test})$"])
plt.subplot(212)
plt.bar(model.classes_, model.predict_proba(test_X)[0])
plt.xlim(-1,2)
plt.gca().xaxis.grid(False)
plt.xticks(model.classes_, ["$P(y=0|x_{test})$", "$P(y=1|x_{test})$"])
plt.title("Conditional probability distribution")
plt.tight_layout()
plt.show()
```

```python
import io
from sklearn.tree import export_graphviz
import pydot
from IPython.core.display import Image
```

```python
def draw_decision_tree(model, feature_names=feature_names):
    dot_buf = io.StringIO()
    export_graphviz(model, out_file=dot_buf, feature_names=feature_names)
    graph = pydot.graph_from_dot_data(dot_buf.getvalue())[0]
    image = graph.create_png()
    return Image(image)
```
